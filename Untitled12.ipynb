```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df= pd.read_csv("C:/Users/PC/Documents/PBJ_Daily_Nurse_Staffing_Q2_2024.csv", encoding="ISO-8859-1",dtype={"PROVNUM": str})
df 
```

```python
column_names = df.columns.tolist()
print(column_names)
```

```python
df=df.drop(df.columns[[9,12,15,18,21,24,27,30]],axis=1)#drops the non ctr, emp
zero_percentage = (df == 0).sum() / len(df) * 100
zero_percentage_sorted = zero_percentage.sort_values(ascending=True)

# Plot horizontal bar chart
plt.figure(figsize=(8, 5))
plt.barh(zero_percentage_sorted.index, zero_percentage_sorted.values, color='skyblue')

# Labels and title
plt.ylabel("Columns")
plt.xlabel("Percentage of Zeros")
plt.title("Percentage of Zeros in Each Column (Decreasing Order)")

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.ticker as mticker

# Get column names and their sums
column_sums = df.sum(numeric_only= True)

# Sort in descending order
column_sums_sorted = column_sums.sort_values()

# Extract sorted names and values


# Plot
plt.figure(figsize=(8, 5))
plt.barh(column_sums_sorted.index, column_sums_sorted.values, color='skyblue')
plt.xlabel("Columns")
plt.ylabel("Sum")
plt.title("Sum of Selected Columns (Sorted)")

# Format x-axis to show full numbers
ax = plt.gca()
formatter = mticker.ScalarFormatter(useOffset=False)
formatter.set_scientific(False)
ax.xaxis.set_major_formatter(formatter)

plt.xticks(rotation=90)
plt.show()
```
